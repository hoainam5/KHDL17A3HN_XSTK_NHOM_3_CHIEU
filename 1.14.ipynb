```python
p_A1=1/4    # A1 là biến cố "Lấy ra từ hộp 1"
p_A2=1/4    # A2 là biến cố "Lấy ra từ hộp 2"
p_A3=1/2    # A3 là biến cố "Lấy ra từ hộp 3"
p_B_A1=7/10 # Phấn trắng lấy ra từ hộp 1"
p_B_A2=4/5  # Phấn trắng lấy ra từ hộp 2"
p_B_A3=21/25 # Phấn trắng lấy ra từ hộp 3"
p_B=p_A1*p_B_A1+p_A2*p_B_A2+p_A3*p_B_A3
print("Xác suất để lấy được viên phấn trắng là: ", p_B)

p_C_A3=4/25     # Phấn vàng lấy từ hộp 3
p_C_A1=3/10     # Phấn vàng lấy từ hộp 1
p_C_A2=1/5      # Phấn vàng lấy từ hộp 2
p_C=p_A1*p_C_A1+p_A2*p_C_A2+p_A3*p_C_A3    # Xác xuất để lấy dược phấn vàng
p_A3_C=(p_A3*p_C_A3)/p_C
print("Xác suất để lấy được viên phấn vàng từ hộp thứ 3 là: ", p_A3_C)
```
