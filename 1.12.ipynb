```python
p_A1=0.25      # A1 là biến cố "Lấy ra từ hộp 1"
p_A2=0.3       # A2 là biến cố "Lấy ra từ hộp 2"
p_A3=0.45      # A3 là biến cố "Lấy ra từ hộp 3"
p_B_A1=0.1     # Phế phẩm lấy ra từ máy 1
p_B_A2=0.2     # Phế phẩm lấy ra từ máy 2
p_B_A3=0.4     # Phế phẩm lấy ra từ máy 3
p_B=p_A1*p_B_A1+p_A2*p_B_A2+p_A3*p_B_A3
print("Xác xuất để chọn được chi tiết phế phẩm là: ", p_B)
p_A1_B=(p_A1*p_B_A1)/p_B
print("Xác suất để lấy chi tiết phế phẩm do máy 1 sản xuất là: ", p_A1_B)
```
