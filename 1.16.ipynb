```python
p_H1=0.3       # Xác suất người đó hút thuốc lá
p_H2=0.7       # Xác suất người đó không hút thuốc lá
p_A_H1=0.6     # Xác suất người đó hút thuốc lá và bị viêm họng
p_A_H2=0.3     # Xác suất người đó hút thuốc lá và không bị viêm họng
p_A=p_H1*p_A_H1+p_H2*p_A_H2
p_H1_A=(p_H1*p_A_H1)/p_A
print("Xác suất người đó bị viêm họng do hút thuốc lá là: ", p_H1_A)

p_B_H1=0.4     # Xác suất người đó hút thuốc lá nhưng không bị viêm họng
p_B_H2=0.7     # Xác suất người đó không hút thuốc lá và không bị viêm họng
p_B=p_H1*p_B_H1+p_H2*p_B_H2
p_H1_B=(p_H1*p_B_H1)/p_B
print("Xác suất để người đó hút thuốc lá nhưng không viêm họng là: ", p_H1_B)
```
