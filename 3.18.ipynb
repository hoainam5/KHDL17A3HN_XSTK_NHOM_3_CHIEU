```python
import math

# Hàm tính tỉ lệ dưới đường cong phân phối chuẩn
def tinh_ty_le_duoi_ddc(X, mu, sigma):
    Z = (X - mu) / sigma
    return (1 + math.erf(Z / math.sqrt(2))) / 2

# Thông số của phân phối chuẩn
mu = 500
sigma = math.sqrt(16)  # căn bậc hai của phương sai

# Tính tỉ lệ cho từng loại trái cây
# Loại 1: trên 505 gam
ty_le_loai_1 = 1 - tinh_ty_le_duoi_ddc(505, mu, sigma)
print("Tỉ lệ loại 1:", ty_le_loai_1)

# Loại 2: từ 495 gam đến 505 gam
ty_le_loai_2 = tinh_ty_le_duoi_ddc(505, mu, sigma) - tinh_ty_le_duoi_ddc(495, mu, sigma)
print("Tỉ lệ loại 2:", ty_le_loai_2)

# Loại 3: dưới 495 gam
ty_le_loai_3 = tinh_ty_le_duoi_ddc(495, mu, sigma)
print("Tỉ lệ loại 3:", ty_le_loai_3)
```
