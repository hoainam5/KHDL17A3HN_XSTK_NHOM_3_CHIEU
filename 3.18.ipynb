```python
import math

# Hm tnh t l di ng cong phn phi chun
def tinh_ty_le_duoi_ddc(X, mu, sigma):
    Z = (X - mu) / sigma
    return (1 + math.erf(Z / math.sqrt(2))) / 2

# Thng s ca phn phi chun
mu = 500
sigma = math.sqrt(16)  # cn bc hai ca phng sai

# Tnh t l cho tng loi tri cy
# Loi 1: trn 505 gam
ty_le_loai_1 = 1 - tinh_ty_le_duoi_ddc(505, mu, sigma)
print("T l loi 1:", ty_le_loai_1)

# Loi 2: t 495 gam n 505 gam
ty_le_loai_2 = tinh_ty_le_duoi_ddc(505, mu, sigma) - tinh_ty_le_duoi_ddc(495, mu, sigma)
print("T l loi 2:", ty_le_loai_2)

# Loi 3: di 495 gam
ty_le_loai_3 = tinh_ty_le_duoi_ddc(495, mu, sigma)
print("T l loi 3:", ty_le_loai_3)
```
