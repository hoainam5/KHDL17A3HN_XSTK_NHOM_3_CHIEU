```python
# Xác định mặt của mỗi con xúc xắc
faces = [1, 2, 3, 4, 5, 6]

# Xây dựng không gian mẫu có thứ tự
sample_space_ordered = [(face1, face2) for face1 in faces for face2 in faces]
print("Không gian mẫu có thứ tự:", sample_space_ordered)
# Xây dựng không gian mẫu không thứ tự
sample_space_unordered = [(face1, face2) for face1 in faces for face2 in faces if face1 <= face2]
print("Không gian mẫu không thứ tự:", sample_space_unordered)
```
