```python
import math

# nh ngha hm tnh t l sn phm phi bo hnh
def ty_le_bao_hanh(X, mu, sigma):
    Z = (X - mu) / sigma
    ty_le = (1 + math.erf(Z / math.sqrt(2))) / 2
    return ty_le

# nh ngha hm tnh li nhun k vng
def loi_nhuan_ki_vong(loinhuan_khong_hong, loinhuan_bi_hong, ty_le_bao_hanh):
    loi_nhuan_ki_vong = (loinhuan_khong_hong * (1 - ty_le_bao_hanh)) + (loinhuan_bi_hong * ty_le_bao_hanh)
    return loi_nhuan_ki_vong

# Tnh t l sn phm phi bo hnh khi X = 980 gi
X = 980
mu = 1000
sigma = 10
ty_le_980 = ty_le_bao_hanh(X, mu, sigma)
print("T l sn phm phi bo hnh khi X = 980 gi:", ty_le_980)

# Tnh li nhun k vng
loinhuan_khong_hong = 150000  # Li nhun t mi sn phm khng hng
loinhuan_bi_hong = -500000     # Li nhun t mi sn phm b hng
loi_nhuan_980 = loi_nhuan_ki_vong(loinhuan_khong_hong, loinhuan_bi_hong, ty_le_980)
print("Li nhun k vng khi X = 980 gi:", loi_nhuan_980)

# Tnh thi gian bo hnh khi li nhun k vng l 45 ngn ng
loi_nhuan_mong_muon = 45000
ty_le_muc_tieu = (loinhuan_khong_hong - loi_nhuan_mong_muon) / (loinhuan_khong_hong - loinhuan_bi_hong)
X_muc_tieu = mu + (sigma * math.sqrt(2) * math.erf(2 * ty_le_muc_tieu - 1))
print("Thi gian bo hnh  t li nhun k vng l 45 ngn ng:", X_muc_tieu)
```
