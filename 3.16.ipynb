```python
import math

# Định nghĩa hàm tính tỉ lệ sản phẩm phải bảo hành
def ty_le_bao_hanh(X, mu, sigma):
    Z = (X - mu) / sigma
    ty_le = (1 + math.erf(Z / math.sqrt(2))) / 2
    return ty_le

# Định nghĩa hàm tính lợi nhuận kì vọng
def loi_nhuan_ki_vong(loinhuan_khong_hong, loinhuan_bi_hong, ty_le_bao_hanh):
    loi_nhuan_ki_vong = (loinhuan_khong_hong * (1 - ty_le_bao_hanh)) + (loinhuan_bi_hong * ty_le_bao_hanh)
    return loi_nhuan_ki_vong

# Tính tỉ lệ sản phẩm phải bảo hành khi X = 980 giờ
X = 980
mu = 1000
sigma = 10
ty_le_980 = ty_le_bao_hanh(X, mu, sigma)
print("Tỉ lệ sản phẩm phải bảo hành khi X = 980 giờ:", ty_le_980)

# Tính lợi nhuận kì vọng
loinhuan_khong_hong = 150000  # Lợi nhuận từ mỗi sản phẩm không hỏng
loinhuan_bi_hong = -500000     # Lợi nhuận từ mỗi sản phẩm bị hỏng
loi_nhuan_980 = loi_nhuan_ki_vong(loinhuan_khong_hong, loinhuan_bi_hong, ty_le_980)
print("Lợi nhuận kì vọng khi X = 980 giờ:", loi_nhuan_980)

# Tính thời gian bảo hành khi lợi nhuận kì vọng là 45 ngàn đồng
loi_nhuan_mong_muon = 45000
ty_le_muc_tieu = (loinhuan_khong_hong - loi_nhuan_mong_muon) / (loinhuan_khong_hong - loinhuan_bi_hong)
X_muc_tieu = mu + (sigma * math.sqrt(2) * math.erf(2 * ty_le_muc_tieu - 1))
print("Thời gian bảo hành để đạt lợi nhuận kì vọng là 45 ngàn đồng:", X_muc_tieu)
```
