```python
import math

# Hàm tính hàm phân phối chuẩn
def phan_phoi_chuan(X, mu, sigma):
    Z = (X - mu) / sigma
    return (1 + math.erf(Z / math.sqrt(2))) / 2

# Thông số của phân phối chuẩn
mu = 250
sigma = 40

# Tính xác suất cho các trường hợp cụ thể
# Xác suất một con bò nặng hơn 300kg
# la diện tích dưới đường cong phân phối chuẩn hoá bên trái giá trị
xac_suat_tren_300 = 1 - phan_phoi_chuan(300, mu, sigma)
print("Xác suất một con bò nặng hơn 300kg:", xac_suat_tren_300)

# Xác suất một con bò nhẹ hơn 175kg
xac_suat_duoi_175 = phan_phoi_chuan(175, mu, sigma)
print("Xác suất một con bò nhẹ hơn 175kg:", xac_suat_duoi_175)

# Xác suất một con bò nằm trong khoảng từ 260kg đến 270kg
xac_suat_trong_khoang_260_270 = phan_phoi_chuan(270, mu, sigma) - phan_phoi_chuan(260, mu, sigma)
print("Xác suất một con bò nằm trong khoảng từ 260kg đến 270kg:", xac_suat_trong_khoang_260_270)
```
