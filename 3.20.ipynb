```python
import math

# Hm tnh hm phn phi chun
def phan_phoi_chuan(X, mu, sigma):
    Z = (X - mu) / sigma
    return (1 + math.erf(Z / math.sqrt(2))) / 2

# Thng s ca phn phi chun
mu = 250
sigma = 40

# Tnh xc sut cho cc trng hp c th
# Xc sut mt con b nng hn 300kg
# la din tch di ng cong phn phi chun ho bn tri gi tr
xac_suat_tren_300 = 1 - phan_phoi_chuan(300, mu, sigma)
print("Xc sut mt con b nng hn 300kg:", xac_suat_tren_300)

# Xc sut mt con b nh hn 175kg
xac_suat_duoi_175 = phan_phoi_chuan(175, mu, sigma)
print("Xc sut mt con b nh hn 175kg:", xac_suat_duoi_175)

# Xc sut mt con b nm trong khong t 260kg n 270kg
xac_suat_trong_khoang_260_270 = phan_phoi_chuan(270, mu, sigma) - phan_phoi_chuan(260, mu, sigma)
print("Xc sut mt con b nm trong khong t 260kg n 270kg:", xac_suat_trong_khoang_260_270)
```
