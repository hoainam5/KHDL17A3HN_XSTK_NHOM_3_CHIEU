```python
# Phép thử 1: Tung một đồng xu cân đối và đồng chất 2 lần liên tiếp
# Kết quả có thể: Sấp (S) hoặc Ngửa (N)
results_1 = ['S', 'N']

# Xây dựng không gian biến cố sơ cấp
sample_space_1 = [(result1, result2) for result1 in results_1 for result2 in results_1]
print("Không gian biến cố sơ cấp của phép thử 1:", sample_space_1)
# Phép thử 2: Một đồng xu được tung ba lần với tình trạng sấp ngửa của các lần tung
# Kết quả có thể: Sấp (S) hoặc Ngửa (N)
results_2 = ['S', 'N']

# Xây dựng không gian biến cố sơ cấp
sample_space_2 = [(result1, result2, result3) for result1 in results_2 for result2 in results_2 for result3 in results_2]
print("Không gian biến cố sơ cấp của phép thử 2:", sample_space_2)
```
